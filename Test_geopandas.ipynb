```python
import geopandas as gpd
import pyproj
import plotly.express as px
```

```python
myshpfile = gpd.read_file('depto/depto.shp')
myshpfile.set_index('DPTO', drop=True, inplace=True)
myshpfile.to_file('depto/depto.geojson', drive='GeoJSON')
myshpfile['nueva'] = 1
myshpfile.head()
```

```python
map_df = myshpfile.to_crs(pyproj.CRS.from_epsg(4326))
map_df.head()
```

```python
fig = px.choropleth(map_df, geojson=map_df.geometry, locations=map_df.index, color="AREA")
fig.update_geos(fitbounds="locations", visible=True)
fig.show()
```
